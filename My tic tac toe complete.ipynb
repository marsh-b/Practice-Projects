```python
from IPython.display import clear_output

#building board
def display_board(board):
    clear_output() 
    
    print('   |   |')
    print(' ' + board[1] + ' | ' + board[2] + ' | ' + board[3])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + board[4] + ' | ' + board[5] + ' | ' + board[6])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + board[7] + ' | ' + board[8] + ' | ' + board[9])
    print('   |   |')
```

```python
#ask for player marker
def marker_input():
    marker = ''
    
    while not (marker == 'X' or marker == 'O'):
        marker = input('Player 1: Do you want to be X or O? ').upper()

    if marker == 'X':
        return ('X', 'O')
    else:
        return ('O', 'X')
```

```python
#place the marker
def place_marker(board, marker, position):
    board[position] = marker
```

```python
#player input
def player_choice(board):
    position = 0
    
    while position not in [1,2,3,4,5,6,7,8,9] or not space_check(board, position):
        position = int(input('Choose your next position: (1-9) '))
        
    return position
```

```python
#turn check
import random

def first_turn():
    if random.randint(1, 2) == 1:
        return 'Player 2'
    else:
        return 'Player 1'
```

```python
#single space check
def space_check(board, position):
    
    return board[position] == ' '
```

```python
#board space check
def full_board_check(board):
    for i in range(1,10):
        if space_check(board, i):
            return False
    return True
```

```python
#win check
def win_check(board,mark):
    return ((board[7] == mark and board[8] == mark and board[9] == mark) or 
(board[5] == mark and board[6] == mark and board[4] == mark) or # across middle
(board[1] == mark and board[2] == mark and board[3] == mark) or # across bottom
(board[7] == mark and board[4] == mark and board[1] == mark) or # Left column
(board[8] == mark and board[5] == mark and board[2] == mark) or # Right column
(board[9] == mark and board[6] == mark and board[1] == mark) or # Middle column
(board[9] == mark and board[1] == mark and board[5] == mark) or # Right bottom to lefttop diag
(board[7] == mark and board[5] == mark and board[3] == mark)) # Left bottom  to right top diag
```

```python
#replay check
def replay():
    
    return input('Do you want to play again? Enter Yes or No: ').lower().startswith('y')
```

```python
# Tic Tac Toe

print('Game start')

#Wipe the board and reset
while True:
    
    theBoard = [' '] * 10 #wipe the baord
    player1_marker, player2_marker = marker_input() #ask for marker input
    turn = first_turn() #decide who goes first
    print(turn + ' will go first.') 
    
    play_game = input('Ready to play? Y/N') #ready to play?
    
    #turn game on
    if play_game.lower()[0] == 'y':
        game_on = True
    else:
        game_on = False

    while game_on:
        if turn == 'Player 1':
            # Player1's turn.
            
            display_board(theBoard) #show board
            position = player_choice(theBoard) #get player move
            place_marker(theBoard, player1_marker, position) #place marker

            if win_check(theBoard, player1_marker): #check for win
                display_board(theBoard)
                print('Player 1 win')
                game_on = False #if win, turn off game
            else:
                if full_board_check(theBoard): #check for full board
                    display_board(theBoard)
                    print('Cat Scratch')
                    break
                else:
                    turn = 'Player 2' #switch over

        else:
            # Player2's turn.
            
            display_board(theBoard) #show board
            position = player_choice(theBoard) #get player move
            place_marker(theBoard, player2_marker, position) #place marker

            if win_check(theBoard, player2_marker): #check for win
                display_board(theBoard)
                print('Player 2 has won!')
                game_on = False #if win, turn off game
            else:
                if full_board_check(theBoard): #check for full board
                    display_board(theBoard)
                    print('Cat Scratch')
                    break
                else:
                    turn = 'Player 1' #switch over

    if not replay():
        break 
```
