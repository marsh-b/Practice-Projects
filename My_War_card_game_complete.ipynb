```python
# Recreating 'War' Card Game- code along with Udemy Class

import random # for shuffling 

#Global Variables
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':11, 'Queen':12, 'King':13, 'Ace':14}
```

```python
class Card: #Create the Cards
    
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
        
    #create readable card for people
    def __str__(self):
        return self.rank + ' of ' + self.suit
    
class Deck: #create the Deck
    
    def __init__(self):
        
        self.all_cards = [] #the deck
        
        for suit in suits: 
            for rank in ranks:
                created_card = Card(suit,rank) #create card object
                
                self.all_cards.append(created_card) #add them to the deck
                
    def shuffle(self): #shuffle the deck
        random.shuffle(self.all_cards)
        
    def deal_one(self): #deal a card from the deck
        return self.all_cards.pop()
    
class Player: #creates the Players' hand
    
    def __init__(self,name):
        self.name = name #player name
        self.all_cards = [] #their hand
    
    def remove_one(self): #remove a card from the top of hand
        return self.all_cards.pop(0)
    
    def add_cards(self,new_cards): #add cards to bottom of hand
        if type(new_cards) == type([]):
            self.all_cards.extend(new_cards) #multiple cards
        else: 
            self.all_cards.append(new_cards) #one card
    
    def __str__(self):
        return f'Player {self.name} has {len(self.all_cards)} cards.' 
```

```python
#game setup

player_one = Player('One') #Assign Player Names
player_two = Player('Two')

new_deck = Deck() #create new deck
new_deck.shuffle() #shuffle

for x in range(26):
    player_one.add_cards(new_deck.deal_one())
    player_two.add_cards(new_deck.deal_one()) #deal 26 cards to each player
    
game_on = True
```

```python
round_num = 0 #round number counts

#game logic
while game_on:

    round_num += 1
    print(f"round {round_num}")
    
    if len(player_one.all_cards) == 0: #check if p1 can play
        print('Player one you lose')
        game_on = False
        break
        
    if len(player_two.all_cards) == 0: #check if p1 can play
        print('Player two you lose')
        game_on = False
        break
        
    player_one_cards = [] #'table' cards
    player_one_cards.append(player_one.remove_one())
    
    player_two_cards = []
    player_two_cards.append(player_two.remove_one())
    
    at_war = True
    
    while at_war:
        
        if player_one_cards[-1].value > player_two_cards[-1].value: #p1 beats p2
            
            player_one.add_cards(player_one_cards)
            player_one.add_cards(player_two_cards)
            
            at_war = False
            
        elif player_two_cards[-1].value > player_one_cards[-1].value: #p2 beats p1
            
            player_two.add_cards(player_one_cards)
            player_two.add_cards(player_two_cards)
            
            at_war = False
            
        else:
            print('WAR')
            
            if len(player_one.all_cards) < 3: #if P1 is out of cards
                print('Player One unable to War, he loses')
                print('P2 wins')
                game_on = False
                break
                
            elif len(player_two.all_cards) < 3: #p2 out of cards
                print('Player Two unable to War, he loses')
                print('P2 wins')
                game_on = False
                break
        
            else:
                for num in range(3): #if both have cards, put 3 on the table and go back to comparison
                    player_one_cards.append(player_one.remove_one())
                    player_two_cards.append(player_two.remove_one())
        
```
