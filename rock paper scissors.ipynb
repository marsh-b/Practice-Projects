```python
import random
```

```python

  
# prints a random value from the list
list1 = ['rock', 'paper', 'scissors']
print(random.choice(list1))
```

```python
your_choice = input('rock paper scissors shoot') 
their_choice = (random.choice(list1)) 

print('YOU chose '+ your_choice + ', THEY chose ' + their_choice)

if your_choice == 'rock' and their_choice == 'scissors':
    print('you win')
if your_choice == 'rock' and their_choice == 'rock':
    print('you tie')
    
if your_choice == 'scissors' and their_choice == 'paper':
    print('you win')
if your_choice == 'scissors' and their_choice == 'scissors':
    print('you tie')
    
if your_choice == 'paper' and their_choice == 'rock':
    print('you win')
if your_choice == 'paper' and their_choice == 'paper':
    print('you tie')
    
else:
        print('you lose')
```
